# Apakah Pendidikan Indonesia Sudah Merata?

Pendidikan adalah hal penting untuk mencetak generasi muda yang unggul di masa depan. Indonesia sebagai negara kepulauan dengan luas wilayah dan jumlah penduduk yang sangat besar, tentu menjadi tantangan tersendiri untuk menyelenggarakan pendidikan yang berkualitas dan merata bagi seluruh rakyatnya. Dalam penelitian ini, saya menggunakan beberapa parameter pendidikan untuk melihat sejauh mana pendidikan indonesia menjangkau masyarakatnya. Dengan metode k-means, saya akan membagi 34 provinsi di Indonesia menjadi beberapa klaster sesuai dengan karakteristik pendidikan masing-masing.

Data yang digunakan adalah data sekunder yang saya peroleh dari publikasi Badan Pusat Statistik maupun dari Kementerian Pendidikan, Kebudayaan, Riset, dan Teknologi. ### Angka Partisipasi 

```python
def trim_axs(axs, N):
    """
    Reduce *axs* to *N* Axes. All further Axes are removed from the figure.
    """
    axs = axs.flat
    for ax in axs[N:]:
        ax.remove()
    return axs[:N]
```

## Data Preprocessing

### Menghapus Data yang Tidak Dibutuhkan

```python
# dalam pembuatan model kita tidak memerlukan variabel provinsi, sehinga dapat kita hapus
df_clean = df.drop('Provinsi', axis=1)
df_clean.head()
```

**Karena variabel provinsi tidak dibutuhkan dalam pembuatan model, sehingga variabel provinsi dapat dihapus.**

### Mendeteksi Data yang Kosong

```python
print('Jumlah data yang kosong pada setiap variabel:')
print(df_clean.isnull().sum())
```

**Pada masing-masing variabel tidak ditemukan data yang kosong.**

### Standarisasi Data 

```python
sc = StandardScaler()
df_trans = sc.fit_transform(df_clean)
df_clean = pd.DataFrame(df_trans,columns=df_clean.columns)
```

```python
df_clean.head()
```

## Modelling

### Elbow Plot

Untuk mencari nilai k yang optimum, perlu dilakukan simulasi. Simulasi dilakukan dengan metode elbow.

```python
# Doing simulation to found the best K number
distortions = []

for i  in range(1,11):
    km = KMeans(n_clusters=i,
                init='k-means++',
                n_init=10,
                max_iter=300,
                random_state=0)
    km.fit(df_clean)
    distortions.append(km.inertia_)

plt.plot(range(1,11),distortions,marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.show()
```

### Clustering

Klasterisasi dilakukan dengan menggunakan metode K-Means, dengan nilai k=4.

```python
km = KMeans(n_clusters=4,
                init='k-means++',
                n_init=10,
                max_iter=300,
                random_state=0)
```

```python
clusters = km.fit(df_clean)
labels = km.labels_
```

```python
df['Clusters'] = labels
col = df.pop('Clusters')

# Insert the column at the desired index
df.insert(1, 'Clusters', col)
```

## Provinsi Berdasarkan Klaster

```python
map = {2:'Tinggi', 1:'Medium', 0:'Rendah', 3:'Sangat Rendah'}
custom = {'Tinggi':0, 'Medium':1, 'Rendah':2, 'Sangat Rendah':3}
hasil['Clusters']=hasil['Clusters'].map(map)
```

```python
hasil1 = hasil.copy()
```

```python
final = hasil1.sort_values(by='Clusters', key=lambda x: x.map(custom)).set_index('Provinsi')
```

```python
def color_categories(val):
    colors = {'Tinggi': 'blue', 'Medium': 'green', 'Rendah': 'orange', 'Sangat Rendah': 'red'}
    font_color = 'white'  # Default font color
    background_color = colors.get(val, 'white')  # Default background color is white
    return f'color: {font_color}; background-color: {background_color}'

# Apply styles
styled_df = final.style.applymap(color_categories, subset=['Clusters'])
```
